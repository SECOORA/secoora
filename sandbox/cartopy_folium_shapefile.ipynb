```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import shapely.geometry as sgeom

import cartopy.crs as ccrs
from cartopy.io import shapereader

ax = plt.axes([0, 0, 1, 1],
              projection=ccrs.LambertConformal())

ax.set_extent([-125, -66.5, 20, 50], ccrs.Geodetic())

kw = dict(resolution='110m', category='cultural',
          name='admin_1_states_provinces')

states_shp = shapereader.natural_earth(**kw)

# To get the effect of having just the states without a map "background"
# turn off the outline and background patches.
ax.background_patch.set_visible(False)
ax.outline_patch.set_visible(False)

shp = shapereader.Reader(states_shp)

SECOORA = ('North Carolina', 'South Carolina', 'Georgia', 'Florida')
for record, state in zip(shp.records(), shp.geometries()):
    name = record.attributes['name']
    if name in SECOORA:
        facecolor = 'DarkOrange'
    else:
        facecolor = 'LightGray'
    ax.add_geometries([state], ccrs.PlateCarree(),
                      facecolor=facecolor, edgecolor='black')
```

```python
import shapefile
from json import dumps

fname = '/home/filipe/.local/share/cartopy/shapefiles/natural_earth/cultural/'
fname += '50m_admin_1_states_provinces_shp.shp'

reader = shapefile.Reader(fname)
fields = reader.fields[1:]

field_names = [field[0] for field in fields]

buffer = []
SECOORA = ('North Carolina', 'South Carolina', 'Georgia', 'Florida')
for sr in reader.shapeRecords():
    atr = dict(zip(field_names, sr.record))
    geom = sr.shape.__geo_interface__
    if 'United States of America' in sr.record[field_names.index('admin')]:
        if sr.record[field_names.index('name')] in SECOORA:
            buffer.append(dict(type="Feature", geometry=geom, properties=atr))
            

with open("secoora.json", "w") as geojson:
    geojson.write(dumps({"type": "FeatureCollection",
                         "features": buffer}, indent=2) + "\n")
```

```python
import folium
import numpy as np

folium.initialize_notebook()

bbox = [-87.40, 24.25, -74.70, 36.70]
lon_center, lat_center = np.array(bbox).reshape(2, 2).mean(axis=0)
secoora = folium.Map(width=750, height=500, zoom_start=5,
                     location=[lat_center, lon_center])

secoora.geo_json(geo_path='secoora.json', fill_color='none', line_color='Orange')
secoora.create_map(path='secoora.html')
secoora
```

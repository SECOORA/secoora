# Access data from the NECOFS (New England Coastal Ocean Forecast System) via OPeNDAP

Demonstration using the NetCDF4-Python library to access velocity data from a triangular grid ocean model (FVCOM) via OPeNDAP, specifying the desired URL, time, layer and lat/lon region of interest.  The resulting plot of forecast velocity vectors over color-shaded bathymetry is useful for a variety of recreational and scientific purposes. 

NECOFS (Northeastern Coastal Ocean Forecast System) is run by groups at the University of Massachusetts Dartmouth and the Woods Hole Oceanographic Institution, led by Drs. C. Chen, R. C. Beardsley, G. Cowles and B. Rothschild. Funding is provided to run the model by the NOAA-led Integrated Ocean Observing System and the State of Massachusetts.

NECOFS is a coupled numerical model that uses nested weather models, a coastal ocean circulation model, and a wave model. The ocean model is a volume-mesh model with horizontal resolution that is finer in complicated regions. It is layered (not depth-averaged) and includes the effects of tides, winds, and varying water densities caused by temperature and salinity changes.

* Model description: http://fvcom.smast.umassd.edu/research_projects/NECOFS/model_system.html
* THREDDS server with other forecast and archive products: http://www.smast.umassd.edu:8080/thredds/catalog.html

### DAP Data URL (MassBay GRID).

```python
import netCDF4
url = 'http://www.smast.umassd.edu:8080/thredds/dodsC/FVCOM/NECOFS/Forecasts/NECOFS_FVCOM_OCEAN_MASSBAY_FORECAST.nc'
nc = netCDF4.Dataset(url)
```

### Get desired time step.

```python
from datetime import datetime, timedelta
# Desired time for snapshot
# ...right now (or some number of hours from now)...
start = datetime.utcnow() + timedelta(hours=18)
# ... or specific time (UTC)
#start = datetime(2013, 3, 2, 15, 0, 0)
```

```python
# Get lon,lat coordinates for nodes (depth)
lon, lat = nc.variables['lon'][:], nc.variables['lat'][:]

# Get lon,lat coordinates for cell centers (depth).
lonc, latc = nc.variables['lonc'][:], nc.variables['latc'][:]

# Get Connectivity array.
nv = nc.variables['nv'][:].T - 1 

# Get depth.
h = nc.variables['h'][:]

# Get time.
time_var = nc.variables['time']
itime = netCDF4.date2index(start, time_var, select='nearest')
dtime = netCDF4.num2date(time_var[itime], time_var.units)
daystr = dtime.strftime('%Y-%b-%d %H:%M')
print(daystr)
```

```python
from matplotlib.tri import Triangulation
tri = Triangulation(lon,lat, triangles=nv)
```

```python
# Get current at layer [0 = surface, -1 = bottom].
ilayer = 0
u = nc.variables['u'][itime, ilayer, :]
v = nc.variables['v'][itime, ilayer, :]
```

### Woods Hole.

```python
import numpy as np

levels = range(-30, 2, 1)  # Depth contours to plot.
lims = [-70.7, -70.6, 41.48, 41.55]
maxvel = 1.0
subsample = 2
```

### Boston harbor.

```python
levels = range(-34, 2, 1)
lims = [-70.97, -70.82, 42.25, 42.35]
maxvel = 0.5
subsample = 3
```

```python
# Find velocity points in bounding box
ind = np.argwhere((lonc >= lims[0]) & (lonc <= lims[1]) & (latc >= lims[2]) & (latc <= lims[3]))
```

```python
np.random.shuffle(ind)
Nvec = int(len(ind) / subsample)
idv = ind[:Nvec]
```

### Tricontourf plot of water depth with vectors on top.

```python
import matplotlib.pyplot as plt

merc_aspect = dict(aspect=(1.0 / np.cos(lat.mean() * np.pi/180.0)))

fig, ax = plt.subplots(figsize=(18, 10), subplot_kw=merc_aspect)
cs = ax.tricontourf(tri, -h, levels=levels, shading='faceted', cmap=plt.cm.gist_earth)
ax.axis(lims)
ax.patch.set_facecolor('0.5')
cbar = fig.colorbar(cs)
cbar.set_label('Water Depth (m)', rotation=-90)
Q = ax.quiver(lonc[idv], latc[idv], u[idv], v[idv], scale=20)
maxstr = '%3.1f m/s' % maxvel
qk = ax.quiverkey(Q, 0.92, 0.08, maxvel, maxstr, labelpos='W')
ax.set_title('NECOFS Velocity, Layer %d, %s UTC' % (ilayer, daystr))
```

```python
# Based on: https://github.com/hetland/octant/blob/master/octant/sandbox/googleearth.py

kml_groundoverlay = '''<?xml version="1.0" encoding="UTF-8"?>
<kml xmlns="http://earth.google.com/kml/2.0">
<Document>
<GroundOverlay>
  <name>__NAME__</name>
  <color>__COLOR__</color>
  <visibility>__VISIBILITY__</visibility>
  <Icon>
    <href>overlay.png</href>
  </Icon>
  <LatLonBox>
    <south>__SOUTH__</south>
    <north>__NORTH__</north>
    <west>__WEST__</west>
    <east>__EAST__</east>
  </LatLonBox>
</GroundOverlay>
<ScreenOverlay>
    <name>Legend</name>
    <Icon>
        <href>legend.png</href>
    </Icon>
    <overlayXY x="0" y="0" xunits="fraction" yunits="fraction"/>
    <screenXY x="0.015" y="0.075" xunits="fraction" yunits="fraction"/>
    <rotationXY x="0.5" y="0.5" xunits="fraction" yunits="fraction"/>
    <size x="0" y="0" xunits="pixels" yunits="pixels"/>
</ScreenOverlay>
</Document>
</kml>
'''

import os
import zipfile
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt


def geo_figure(lon, lat, prop, **kwargs):
    name = kwargs.pop('name', 'overlay')
    color = kwargs.pop('color', '9effffff')
    visibility = str(kwargs.pop('visibility', 1))
    kmzfile = kwargs.pop('kmzfile', 'overlay.kmz')
    pixels = kwargs.pop('pixels', 1024)  # Pixels of the maximum dimension.
    units = kwargs.pop('units', '')
    vmax = kwargs.pop('vmax', prop.max())
    kwargs['vmax'] = vmax
    vmin = kwargs.pop('vmin', prop.min())
    kwargs['vmin'] = vmin

    geo_aspect = np.cos(lat.mean() * np.pi/180.0)
    xsize = lon.ptp() * geo_aspect
    ysize = lat.ptp()

    aspect = ysize / xsize
    if aspect > 1.0:
        figsize = (10.0 / aspect, 10.0)
    else:
        figsize = (10.0, 10.0 * aspect)

    fig = plt.figure(figsize=figsize,
                     facecolor=None,
                     frameon=False,
                     dpi=pixels//10)
    ax = fig.add_axes([0, 0, 1, 1])
    
    #pc = ax.pcolor(lon, lat, prop, **kwargs)
    #ax.set_xlim(lon.min(), lon.max())
    #ax.set_ylim(lat.min(), lat.max())
    
    pc = ax.tricontourf(tri, -h, levels=levels, shading='faceted', cmap=plt.cm.gist_earth)
    ax.axis(lims)
    ax.patch.set_facecolor('0.5')
    #cbar = fig.colorbar(cs)
    #cbar.set_label('Water Depth (m)', rotation=-90)
    Q = ax.quiver(lonc[idv], latc[idv], u[idv], v[idv], scale=20)
    maxstr = '%3.1f m/s' % maxvel
    qk = ax.quiverkey(Q, 0.92, 0.08, maxvel, maxstr, labelpos='W')
    ax.set_title('NECOFS Velocity, Layer %d, %s UTC' % (ilayer, daystr))

    ax.set_axis_off()
    fig.savefig('overlay.png')

    f = zipfile.ZipFile(kmzfile, 'w')
    f.writestr('overlay.kml',
               kml_groundoverlay.replace('__NAME__', name)
                                .replace('__COLOR__', color)
                                .replace('__VISIBILITY__', visibility)
                                .replace('__SOUTH__', str(lat.min()))
                                .replace('__NORTH__', str(lat.max()))
                                .replace('__EAST__', str(lon.max()))
                                .replace('__WEST__', str(lon.min())))
    f.write('overlay.png')
    os.remove('overlay.png')

    fig = plt.figure(figsize=(1.0, 4.0),
                     facecolor=None,
                     frameon=False)
    ax = fig.add_axes([0.0, 0.05, 0.2, 0.9])
    cb = fig.colorbar(pc, cax=ax)
    cb.set_label(units, color='0.9')
    for lab in cb.ax.get_yticklabels():
        plt.setp(lab, 'color', '0.9')

    plt.savefig('legend.png')
    f.write('legend.png')
    os.remove('legend.png')
    f.close()
```

```python
geo_figure(lon, lat, h, kmzfile='FVCOM.kmz',
           name='FVCOM_depth_and_velocity',
           units='Water Depth (m)')
```

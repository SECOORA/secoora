```python
import iris
import pytz
from datetime import datetime, timedelta

from utilities import CF_names

stop = datetime(2015, 5, 8, 12)
stop = stop.replace(tzinfo=pytz.utc)
start = stop - timedelta(days=7)

bbox = [-87.40, 24.25, -74.70, 36.70]

name_list = CF_names['water_surface_height_above_reference_datum']

units = iris.unit.Unit('m')
```

```python
import warnings
from utilities import quick_load_cubes, proc_cube, get_surface

sabgom = ("http://omgsrv1.meas.ncsu.edu:8080/thredds/dodsC/fmrc/sabgom/"
          "SABGOM_Forecast_Model_Run_Collection_best.ncd")

useast = ("http://omgsrv1.meas.ncsu.edu:8080/thredds/dodsC/fmrc/us_east/"
          "US_East_Forecast_Model_Run_Collection_best.ncd")

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    sabgom = quick_load_cubes(sabgom, name_list, callback=None, strict=True)
    sabgom = proc_cube(sabgom, bbox=bbox, time=(start, stop), units=units)
    sabgom = get_surface(sabgom)
    
    useast = quick_load_cubes(useast, name_list, callback=None, strict=True)
    useast = proc_cube(useast, bbox=bbox, time=(start, stop), units=units)
    useast = get_surface(useast)
```

```python
from utilities import get_nearest_water, make_tree

obs = dict(lon=-78.9106334839, lat=33.621892401)

tree, sabgom_lon, sabgom_lat = make_tree(sabgom)
kw = dict(k=10, max_dist=0.04, min_var=0.01)
sabgom_series, dist, sabgom_idx = get_nearest_water(sabgom, tree, obs['lon'], obs['lat'], **kw)

print('Distance (degrees): {}'.format(dist))
print('Indices: {!r}'.format(sabgom_idx))
```

```python
tree, useast_lon, useast_lat = make_tree(useast)
kw = dict(k=10, max_dist=0.04, min_var=0.01)
useast_series, dist, useast_idx = get_nearest_water(useast, tree, obs['lon'], obs['lat'], **kw)

print('Distance (degrees): {}'.format(dist))
print('Indices: {!r}'.format(useast_idx))
```

```python
%matplotlib inline

import cartopy.crs as ccrs
import matplotlib.pyplot as plt
from cartopy.feature import NaturalEarthFeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER


states = NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                             name='admin_1_states_provinces_shp')

def make_map(projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(8, 6),
                           subplot_kw=dict(projection=projection))
    ax.coastlines(resolution='50m', zorder=0)
    gl = ax.gridlines(draw_labels=True)
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
import numpy.ma as ma


c = useast[-1, ...]  # Latest temperature.
c.data = ma.masked_invalid(c.data)

lon = c.coord(axis='X').points
lat = c.coord(axis='Y').points

extent = (lon.min(), lon.max(),
          lat.min(), lat.max())
```

```python
from oceans import cm

from matplotlib.ticker import MaxNLocator
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes



fig, ax = make_map()
ax.set_extent(extent)
cs = ax.pcolormesh(useast_lon, useast_lat, c.data, cmap=cm.rscolmap, alpha=0.5, zorder=1)
ax.plot(useast_lon[useast_idx], useast_lat[useast_idx], marker='>', color='purple', label='SABGOM', zorder=3, linestyle='none')
ax.plot(sabgom_lon[sabgom_idx], sabgom_lat[sabgom_idx], marker='<', color='darkgreen', label='USEAST', zorder=3, linestyle='none')
ax.plot(obs['lon'], obs['lat'], marker='o', color='crimson', label='Springmaid Pier, SC', zorder=2, linestyle='none', alpha=0.5)
ax.add_feature(states, edgecolor='gray')
leg = ax.legend(numpoints=1, loc='upper left')

# Zoom 1.
axins = zoomed_inset_axes(ax, 80, loc=5)
axins.plot(useast_lon[useast_idx], useast_lat[useast_idx], marker='>', color='darkgreen', label='SABGOM', zorder=3, linestyle='none', alpha=0.5)
axins.plot(sabgom_lon[sabgom_idx], sabgom_lat[sabgom_idx], marker='<', color='purple', label='USEAST', zorder=3, linestyle='none')
axins.plot(obs['lon'], obs['lat'], marker='o', color='crimson', label='Springmaid Pier, SC', zorder=2, linestyle='none', alpha=0.5)
axins.axis([-78.95, -78.9, 33.6, 33.65])
axins.xaxis.tick_top()
axins.yaxis.tick_left()
mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

axins.axes.get_xaxis().set_ticks([])
axins.axes.get_yaxis().set_ticks([])

fig.savefig("Springmaid_Pier_SC_points.png", transparent=False)
```

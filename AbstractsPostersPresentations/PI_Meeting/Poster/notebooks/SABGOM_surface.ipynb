```python
import warnings

import iris
from iris.fileformats.cf import reference_terms


iris.FUTURE.netcdf_promote = True
reference_terms['ocean_s_coordinate_g1'] = ['eta', 's']


url = ('http://omgsrv1.meas.ncsu.edu:8080/thredds/dodsC/fmrc/sabgom/'
       'SABGOM_Forecast_Model_Run_Collection_best.ncd')

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    cubes = iris.load_raw(url)
```

```python
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt

from oceans import cm
from utilities import time_coord, z_coord

import cartopy.crs as ccrs
from cartopy.feature import NaturalEarthFeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

from iris import unit

states = NaturalEarthFeature(category='cultural', scale='50m', facecolor='none',
                             name='admin_1_states_provinces_shp')


def value(v):
    try:
        v = float(v)
    except ValueError:
        v = None
    return v


def make_map(projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(8, 6),
                           subplot_kw=dict(projection=projection))
    ax.coastlines(resolution='50m')
    gl = ax.gridlines(draw_labels=True)
    gl.xlabels_top = gl.ylabels_right = False
    gl.xformatter = LONGITUDE_FORMATTER
    gl.yformatter = LATITUDE_FORMATTER
    return fig, ax


def plt_model(name='sea_water_potential_temperature',
              time=0, lev=0, vmin="None", vmax="None"):
    cube = cubes.extract_strict(name)
    if cube.ndim == 3:
        cube = cube[time, ...]
    elif cube.ndim == 4:
        cube = cube[time, lev, ...]
    else:
        print('Unhandled shape {!r}'.format(cube.shape))
    cube.data = ma.masked_invalid(cube.data)
    lon = cube.coord(axis='X').points
    lat = cube.coord(axis='Y').points
    t = time_coord(cube)
    time = t.units.num2date(t.points)[0]

    # Plotting.
    vmin = value(vmin)
    vmax = value(vmax)

    fig, ax = make_map()
    ax.set_extent([lon.min(), lon.max(), lat.min(), lat.max()])
    kw = dict(cmap=cm.rscolmap, alpha=0.5)  # Make cmap a choice.
    cs = ax.pcolormesh(lon, lat, cube.data, vmin=vmin, vmax=vmax)
    model = cube.attributes.get('title', None)
    
    long_name = cube.long_name
    units = cube.units.format(unit.UT_UTF8)
    lim = u'vmin: {} vmax: {}'.format(np.floor(np.nanmin(cube.data)),
                                      np.ceil(np.nanmax(cube.data)))
    title = u'{}\n{} at {}'.format(lim, long_name, time)
    if model:
        title = u'{}\n{}'.format(title, model)
    ax.set_title(title)
    ax.add_feature(states, edgecolor='gray')
    kw = dict(orientation='vertical', extend='both', shrink=0.85)
    cbar = fig.colorbar(cs, **kw)
    cbar.ax.set_title(u'{}'.format(units.decode('utf8')))
```

```python
%matplotlib inline

from IPython.html import widgets

names = [cube.name() for cube in cubes if cube.ndim >= 3]
nt = cubes.extract_strict('forecast_period').shape[0]
nz = min([cube.shape[0] for cube in cubes.extract('ocean_s_coordinate_g1')])

w = widgets.interact(plt_model, name=names,
                     vmin="None", vmax="None",
                     time=(0, nt-1, 1),
                     lev=(0, nz-1, 1))
```
